## Задание 3

В игре Plants & Gardens каждый месяц проводятся тематические события, ограниченные по времени. В них игроки могут получить уникальные предметы для сада и персонажей, дополнительные монеты или бонусы. Для получения награды требуется пройти ряд уровней за определенное время. С помощью каких метрик можно оценить результаты последнего прошедшего события?

Предположим, в другом событии мы усложнили механику событий так, что при каждой неудачной попытке выполнения уровня игрок будет откатываться на несколько уровней назад. Изменится ли набор метрик оценки результата? Если да, то как?

В зависимости от целей проведения таких событий и от глобальных целей компании, а также от жизненного цикла, на котором компания находится метрики могут быть разные.

В стадии, когда компания ценит свою аудиторию и стремится ее развивать, для игры будет важно не только ценность, которую игрок приносит во время ивента, но и его долгосрочное удержание. Таким образом, я бы предложила следующие показатели, разделив их по системам уровней.

### Целевые:
- Выручка (общий показатель выручки за этот период), денежные средства могут приходить, как от просмотра рекламы пользователем, так и от покупок внутри приложения.

Если игра монетизируется также с помощью покупок каких-либо инструментов, которые облегчают прохождение игры:
- ARPU - средняя выручка на одного пользователя (за весь период проведения ивента)
- ARPPU - средняя выручка с одного платящего пользователя
- Percentage of Paying Users - доля платящей аудитории

### Опережающие:
- Sticky Factor = DAU/MAU (или WAU если ивент длится неделю)
- Average session length - средняя продолжительность сессии игрока за время ивента
- Sessions - среднее количество сессий на игрока за время ивента    .
   
### Guardrail: 
- Процент, завершивших ивент
- Среднее кол-во заработанных наград за период (может влиять на дальнейшую лояльность игрока, подкреплять его интерес к игре)


При усложнении механики, когда игрок при неудаче возвращается на несколько уровней назад, общий набор метрик изменится не должен, но нужно внимательнее следить за опережающими показателями, так как такое усложнение может вызывать фрустрацию, и если эти показатели снизятся, это отразится на целевых показателях ивента, а также на дальнейшую лояльность. Но если найти условный баланс, и при условии, что игра монетезируется с помощью внутренних покупок, это можент увеличить денежные показатели, так как игроки после нескольких неудач купят какой-то инструмент, чтобы пройти уровень легче.
